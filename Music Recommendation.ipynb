```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
from sklearn.model_selection import train_test_split
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
def preprocess_text(text):
    text = text.lower()
    # Remove special characters and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    return text

df['content'] = df['content'].apply(preprocess_text)
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
def get_similarities(song_name, data):
    text_array1 = song_vectorizer.transform(data[data['track_name']==song_name]['track_genre']).toarray()
    num_array1 = data[data['track_name']==song_name].select_dtypes(include=np.number).to_numpy()

    sim = []
    for idx, row in data.iterrows():
        track_name = row['track_name']

        text_array2 = song_vectorizer.transform(data[data['track_name']==track_name]['track_genre']).toarray()
        num_array2 = data[data['track_name']==track_name].select_dtypes(include=np.number).to_numpy()

        text_sim = cosine_similarity(text_array1, text_array2)[0][0]
        num_sim = cosine_similarity(num_array1, num_array2)[0][0]
        sim.append(text_sim + num_sim)

    return sim
```

```python
def predict_emotion(text, model, vectorizer, top_n=3):
    # Transforming the input text using the vectorizer
    text_vectorized = vectorizer.transform([text])
    
    # Getting the predicted probabilities for each class
    predicted_probabilities = model.predict_proba(text_vectorized)
    
    # Getingt the indices of the top_n emotions
    top_indices = predicted_probabilities[0].argsort()[-top_n:][::-1]  
    
    # Getting the corresponding emotion labels
    emotion_labels = model.classes_  
    top_emotions = emotion_labels[top_indices]  
    return top_emotions
```

```python
def recommend_songs_based_on_text(text):
    predicted_emotions = predict_emotion(text, model, vectorizer)
    
    matching_genres = set()
    for emotion in predicted_emotions:
        for genre, details in genre_emotion_map.items():
            if emotion in details['emotions']:
                matching_genres.add(genre)

    if not matching_genres:
        print("No matching genres found for the predicted emotions.")
        return

    num_songs = int(input("Number of Songs to Recommendd: "))

    recommended_songs = pd.DataFrame()
    for genre in matching_genres:
        songs_in_genre = data[data['track_genre'] == genre][['track_name', 'artists', 'track_genre', 'popularity']]
        recommended_songs = pd.concat([recommended_songs, songs_in_genre])

    recommended_songs = recommended_songs.sort_values(by='popularity', ascending=False).drop_duplicates()

    if recommended_songs.empty:
        print("No songs found for the matched genres.")
    else:
        print("Recommended Songs based on the analyzed emotions:")
        display(recommended_songs.head(num_songs))
```
